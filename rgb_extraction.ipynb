```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in color (BGR format by default)
image = cv2.imread('DR2.png')

# Check if image loaded successfully
if image is None:
    raise ValueError("Image not found. Please check the file path.")

# Set blue channel (channel 0) to zero
image_no_blue = image.copy()
image_no_blue[:, :, 0] = 0  # Blue channel in OpenCV is index 0

# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_no_blue_rgb = cv2.cvtColor(image_no_blue, cv2.COLOR_BGR2RGB)

# Display original and modified image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Image without Blue Channel")
plt.imshow(image_no_blue_rgb)
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Save the modified image
cv2.imwrite('DR2_no_blue.png', image_no_blue)
```
